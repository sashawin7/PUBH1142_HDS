## Study Questions

### Question 1
**What is the difference between a spreadsheet and a database?**

> Spreadsheets are smaller-scale, cell-based, tabular displays of row and column datasets, and they are usually limited in built-in methods to perform analysis beyond basic statistical testing and data visualizations. 


> Databases are larger-scale, structured collections and are usually compiled of more disparate types and categories of complex data compared to spreadsheets. 


> One mainly interacts with spreadsheets by viewing their cell make-up and performing point and click analysis. Whereas for databases, ones usually interacts with them by searching the front and downloading data, or using an API or associated software to pull data into a different environment.


### Question 2
**What are the advantages and disadvantages of using a spreadsheet for health data management?**

> Spreadsheets are at a great advantage when one is dealing with a smaller quantity of data and needs less ability to manipulate and deal with complex-structured data. They are easier to create, manage, and use.

> The major disadvantages are that they are less equipped to deal with large/complex datasets, they are less secure, and they are limited by the capabilities of the software that they are integrated into. The security and storage/scalability problems mean that it is usually not feasible for organizations or research project groups to use spreadsheets as their main mechanism for storing data.

### Question 3
**What are the advantages and disadvantages of using a database for health data management?**

> The major advantages of databases are their inherent capacity to store large/complex datasets, their built in or addable security measures, the multi-user support of the systems, and the options of better and more dynamic data manipulation techniques. This makes them well-fit to serve as the data storage systems for most large-scale collections of health data.

> The biggest disadvantage of databases is their higher skill floor for creation, management, and use. In order to build and maintain them, one needs to be proficient in database design, which is quite complex. Sometimes their accessibility can be augmented if the system has a significantly built out front end, but they are still usually most optimally accessed when one uses SQL or an API.
